```python
import json
import html
from IPython.core.display import display, HTML
```

```python
with open('<transcript file path>') as json_file:
    text_basis = json.load(json_file)
```

```python
text_basis
```

```python
sentences = text_basis['sentences']
```

```python
text = ''
current_speaker = ''
blocks = []
for sentence in text_basis['sentences']:
    for token in sentence['tokens']:
        if current_speaker == '':
            current_speaker = token['speakerTag']
        elif current_speaker != token['speakerTag']:
            temp_element = (text, current_speaker)
            blocks.append(temp_element)
            text = ''
            current_speaker = token['speakerTag']
        text += token['originalString'] + ' '
temp_element = (text, current_speaker)
blocks.append(temp_element)
for block in blocks:
    print(block)
```

```python
snt_spk_pair = [(sentence['originalSentence'], sentence['tokens'][0]['speakerTag']) for sentence in text_basis['sentences']]
```

```python
def bubble_sentences_labeled(sentence_speaker_pairs):
    out_markup = ''

    for sentence_pair in sentence_speaker_pairs:
        out_markup += '<div class=' + sentence_pair[1] + '_wrapper>'
        out_markup += '<h1><span>' + sentence_pair[1] + '</span></h1></div>'
        out_markup += '<div class=\'' + sentence_pair[1] + '\'>'
        out_markup += '<p>' + sentence_pair[0] + '</p> </div>'
    
    return out_markup

def visualize_imessage_labeled(sentences_and_speakers):
    
    markup_text = """<html><head><style>
    
        div.clinician_wrapper {
            float: right;
            text-align: right;
            clear: both;
            max-width: 300px;
            background-color: white;
            color: black;
            font-family: "SanFrancisco", sans-serif;
            margin: 0.7rem 0;
            width: fit-content;
          }
          div.patient_wrapper {
            float: left;
            text-align: left;
            clear: both;
            max-width: 300px;
            background-color: white;
            color: black;
            font-family: "SanFrancisco", sans-serif;
            margin: 0.7rem 0;
            width: fit-content;
          }

          div.clinician_wrapper h1{
            text-align: left;
            margin-top: -10px;
            height: 15px;
            line-height: 20px;
            font-size: 12px;
            color: black;
          }
          div.patient_wrapper h1{
            text-align: left;
            margin-top: -10px;
            height: 15px;
            line-height: 20px;
            font-size: 12px;
            color: black;
          }

          div.clinician {
            float: right;
            text-align: left;
            clear: both;
            max-width: 300px;
            border-radius: 8px;
            background-color: #0099ff;
            color: #e0ecff;
            font-family: "SanFrancisco", sans-serif;
            width: fit-content;
          }
          div.patient {
            float: left;
            text-align: left;
            clear: both;
            max-width: 300px;
            border-radius: 8px;
            background-color: gray;
            color: black;
            font-family: "SanFrancisco", sans-serif;
            margin-top: 100px;
            margin: 0.7rem 0;
            width: fit-content;
          }

          div.patient p{margin:10px;}
          div.clinician p{margin:10px;}
        </style>
        </head>
        <body>
        """
    markup_text += bubble_sentences_labeled(sentences_and_speakers)
    markup_text += '</body>' + '</html>'

    display(HTML(markup_text))
    
    return markup_text
```

```python
def bubble_sentences(sentence_speaker_pairs):
    out_markup = ''

    for sentence_pair in sentence_speaker_pairs:
        out_markup += '<div class=' + sentence_pair[1] + '>'
        out_markup += '<p>' + sentence_pair[0] + '</p>' + '</div>'
    
    return out_markup

def visualize_imessage(sentences_and_speakers):
    
    markup_text = """<html><head><style>
        div.clinician {
          text-align: left;
          float: right;
          clear: both;
          max-width: 300px;
          border-radius: 8px;
          padding-top: 0px;
          padding-left: 4px;
          background-color: #0099ff;
          color: #e0ecff;
          font-family: "SanFrancisco", sans-serif;
          margin: 0.25rem 0;
          width: fit-content;
        }
        div.clinician p{margin:9px;}

        div.patient {
          text-align: left;
          float: left;
          clear: both;
          max-width: 300px;
          border-radius: 8px;
          padding-left: 4px;
          background-color: #d6d6d6;
          font-family: "SanFrancisco", sans-serif;
          margin-top: 100px;
          margin: 0.25rem 0;
          width: fit-content;
        }
        div.patient p{margin:9px;}

        </style>
        </head>
        <body>
        """
    markup_text += bubble_sentences(sentences_and_speakers)
    markup_text += '</body>' + '</html>'
    
    display(HTML(markup_text))
    
    return markup_text
```
